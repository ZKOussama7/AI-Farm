```python
# def safefile(var1):
#         # Name of the file to read
#     # Read the contents of the file
#     file_contents=""
#     good_content=""
#     with open("./"+str(var1), "r") as file:
#         file_contents=file.read().replace('\x00','')
#     file_contents=file_contents.split('\n')

#     for line in file_contents:
#         parts=line.split(',')
#         try:
#             tt= eval(parts[0])
#             T= inrange(eval(parts[1]),10,60)
#             H= inrange(eval(parts[2]),10,100)
#             L= inrange(eval(parts[3]),0,100)
#             W= inrange(eval(parts[4]),0,50)
#             S1= inrange(eval(parts[5]),0,100)
#             S2= inrange(eval(parts[5]),0,100)
#             S3= inrange(eval(parts[5]),0,100)
#             S4= inrange(eval(parts[5]),0,100)
        
#             if len(parts)==9 and tt>1683933827742 and T and H and L and W and S1 and S2 and S3 and S4:
#                 good_content="\n\n".join((good_content,line))
#         except:
#             pass
        
#     with open("./"+str(var1), "w") as file:
#         file.write(good_content)
# def secfile(var1):
#         # Name of the file to read
#     # Read the contents of the file
#     file_contents=""
#     good_content=""
#     with open("./"+str(var1), "r") as file:
#         file_contents=file.read()
#     file_contents=file_contents.split('\n')
#     ltt=None
#     for line in file_contents:
#         if line=="":
#             continue
#         parts=line.split(',')
#         try:
#             tt= eval(parts[0])
#             if ltt is None:
#                 ltt=tt
#                 continue
#             if(tt-ltt<100):
#                 continue        
#             good_content="\n".join((good_content,line))
#         except:
#             pass
        
#     with open("./"+str(var1), "w") as file:
#         file.write(good_content)
def sendfile(var1):
    # URL of the server to send the file contents to
    file_contents=""
    url = "http://localhost:8080/api/send"
    with open("./"+str(var1), "r") as file:
        file_contents=file.read()
    # Add "File:" prefix to the file contents
    file_contents_with_prefix = f"File:{file_contents}"
    header={'Content-Type':"text/data"}
    # Send the file contents to the server using the POST method
    response = rq.post(url, data=file_contents_with_prefix,headers=header)

    # Print the response from the server
    print(response.text)
```
